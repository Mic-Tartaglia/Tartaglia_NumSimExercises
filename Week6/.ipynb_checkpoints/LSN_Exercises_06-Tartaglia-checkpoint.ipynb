In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

```python
x = np.arange(0.5,2.1,0.1)
ene = np.loadtxt("es6.final/output.ene.final")
ene_err = np.loadtxt("es6.final/output.ene.error")

ene_gibbs = np.loadtxt("es6.gibbs/1sim/output.ene.final")
ene_err_gibbs = np.loadtxt("es6.gibbs/1sim/output.ene.error")


fig = plt.figure(figsize=(15,8))
plt.errorbar(x,ene,yerr = ene_err,color = 'blue')
plt.errorbar(x,ene_gibbs,yerr = ene_err_gibbs,color  = 'red')

T = np.linspace(0.2,2.5,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e,color = 'green')
plt.title('Internal Energy in Ising 1D Model vs Temperature',fontsize = 20)
plt.xlabel('Temperature T',fontsize = 15)
plt.ylabel('U(T)',fontsize = 15)
plt.legend(['Analytical Model','Numeric result - Metropolis', 'Numeric result - Gibbs'],fontsize = 15)
plt.grid()

plt.show()
```

```python
x = np.arange(0.5,2.1,0.1)
heat = np.loadtxt("es6.final/output.heat.final")
heat_err = np.loadtxt("es6.final/output.heat.error")

heat_gibbs = np.loadtxt("es6.gibbs/1sim/output.heat.final")
heat_err_gibbs = np.loadtxt("es6.gibbs/1sim/output.heat.error")


fig = plt.figure(figsize=(15,8))
plt.errorbar(x,2*heat,yerr = heat_err,color = 'blue')
plt.errorbar(x,2*heat_gibbs,yerr = heat_err_gibbs, color = 'red')


heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat,color = 'green')
plt.title('Heat Capacity in Ising 1D Model vs Temperature',fontsize = 20)
plt.xlabel('Temperature T',fontsize = 15)
plt.ylabel('C(T)',fontsize = 15)
plt.legend(['Analytical Model','Numeric result - Metropolis','Numeric result - Gibbs'],fontsize = 15)
plt.grid()

plt.show()
```

**The results are in accordance with the exact solution until the temperature reaches below about 1.0, then the numeric results detach. This is probably due to the probability of flipping a spin being too low, breaking the ergodicity of the system**

```python
x = np.arange(0.5,2.1,0.1)
mag = np.loadtxt("es6.final/h002_sim/output.mag.final")
mag_err = np.loadtxt("es6.final/h002_sim/output.mag.error")

mag_gibbs = np.loadtxt("es6.gibbs/2sim/output.mag.final")
mag_gibbs_err = np.loadtxt("es6.gibbs/2sim/output.mag.error")

fig = plt.figure(figsize=(15,8))
#plt.errorbar(x,mag,yerr = np.arange(0.1,0.2,0.1/16))
plt.errorbar(x,mag,yerr = mag_err,color = 'blue')
plt.errorbar(x,mag_gibbs,yerr = mag_gibbs_err,color = 'red')

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M,color = 'green')
plt.title('Magnetization in Ising 1D Model vs Temperature (h = 0.02)',fontsize = 20)
plt.xlabel('Temperature T',fontsize = 15)
plt.ylabel('M(T)',fontsize = 15)
plt.legend(['Analytical Model','Numeric result - Metropolis','Numeric result - Gibbs'],fontsize = 15)
plt.grid()


plt.show()
```

**Again the data is not very accurate at low temperature, although in a less evident way.**
